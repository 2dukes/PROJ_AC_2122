Bivariate Analysis

Finding some kind of empirical relationship between two variables. Specifically the dependent vs independent variables.

```python
# Loading Packages

from IPython.display import display
import pandas as pd                 
import numpy as np  #for mathematical calculation
import seaborn as sns   #for data visualization
import matplotlib.pyplot as plt # for plotting graphs

merge_train_data = pd.read_csv("./preprocessed/mergedTrainData.csv", sep=",", index_col=0)

def plot_with_count(x, hue) :
    sns.displot(merge_train_data, x=x, hue=hue, palette=sns.color_palette("bright", 2))

def plot_with_percentage(x, hue):
    sns.displot(merge_train_data, x=x, stat='percent', hue=hue, palette=sns.color_palette("bright", 2))
```

```python
# Gender vs. Loan Status

plot_with_count("gender", "status")
```

```python
# Age Group vs. Loan Status

plot_with_count("age_group", "status")
```

```python
# Payments vs. Loan Status

plot_with_count("payments", "status")
```

```python
# Region vs. Loan Status

plot_with_count("region", "status")
```

```python
# Age_Group vs. Loan Status

plot_with_count("age_group", "status")
```
